```python
import json
import pandas as pd
import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
from nltk.tokenize import sent_tokenize
import nltk

# Ensure you have the necessary nltk data
nltk.download('punkt')

# Load JSON data
# link to (/Users/areejnasser/Desktop/ontology-based classification/PhD-2024/Toward Obsessive-Compulsive Disorder Classification System/Ontology/Ontology-Enrichment/Contextual-similarity/Embedding-Similarity-Task)
with open('/Users/areejnasser/Desktop/ontology-based classification/PhD-2024/Toward Obsessive-Compulsive Disorder Classification System/Ontology/Ontology-Enrichment/Contextual-similarity/Embedding-Similarity-Task/OCD2(updated).json', 'r') as file:
    data = json.load(file)

# Assuming the JSON structure contains a key 'TrainingData' with a list of texts
texts = data.get('TrainingData', [])

# Verify the structure and type of each text entry
for i, text in enumerate(texts):
    if not isinstance(text, str):
        print(f"Entry {i} is not a string: {text}")
    else:
        print(f"Entry {i} is a valid string.")

# Tokenize the texts into sentences and words
sentences = []
for text in texts:
    if isinstance(text, str):  # Ensure the text is a string before processing
        for sentence in sent_tokenize(text):
            sentences.append(simple_preprocess(sentence))

print(f"Total sentences for training: {len(sentences)}")
```

```python
import nltk
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Ensure you have the necessary nltk data
nltk.download('punkt')

# Train Word2Vec models with different vector sizes
vector_sizes = [100, 200, 300]
models = {}

for size in vector_sizes:
    model = Word2Vec(sentences, vector_size=size, window=5, min_count=1, workers=4)
    model.save(f"word2vec_ocd_{size}.model")
    models[size] = model
    print(f"Word2Vec model with vector size {size} trained and saved.")


```

```python
# Words to find similar terms for
target_words = ["intrusive", "obsession", "compulsion", "urge"]

# Function to find and print top 10 similar terms
def find_similar_terms(model, words):
    similar_terms = {}
    for word in words:
        try:
            similar = model.wv.most_similar(word, topn=10)
            similar_terms[word] = similar
        except KeyError:
            # Word not in vocabulary
            print(f"Word '{word}' not found in vocabulary.")
            similar_terms[word] = []
    return similar_terms

# Evaluate each model and print similar terms
for size in vector_sizes:
    print(f"\nTop 10 similar terms for vector size {size}:")
    model = models[size]
    similar_terms = find_similar_terms(model, target_words)
    for word, similar in similar_terms.items():
        print(f"\nWord: {word}")
        for similar_word, similarity in similar:
            print(f"  {similar_word}: {similarity:.4f}")
```

```python
from gensim.models.phrases import Phrases, Phraser

# Build the bi-grams model
bigram = Phrases(sentences, min_count=5, threshold=5)
bigram_mod = Phraser(bigram)

# Apply the bi-grams model to the sentences
sentences_bigrams = [bigram_mod[sent] for sent in sentences]

# Print a few example sentences to verify the bi-grams
print("Example bi-grammed sentences:")
for i in range(3):
    print(sentences_bigrams[i])
```

```python
# Train Word2Vec model on bi-grammed sentences
model = Word2Vec(sentences_bigrams, vector_size=100, window=5, min_count=1, workers=4)
model.save("word2vec_ocd_bigrams.model")

print("Word2Vec model trained and saved.")
```

```python

# Load the trained Word2Vec bi-gram model
model = Word2Vec.load("word2vec_ocd_bigrams.model")

# Define the input phrases
input_phrases = ["intrusive thought", "intrusive urge", "compulsive behavior", "hoarding obsession", "obsessive thought"]

# Preprocess the input phrases to match the bi-grams format
def preprocess_phrase(phrase):
    return "_".join(simple_preprocess(phrase))

preprocessed_phrases = [preprocess_phrase(phrase) for phrase in input_phrases]

# Find the top 10 similar terms for each preprocessed phrase
similar_terms = {}
for phrase in preprocessed_phrases:
    try:
        # Get the most similar terms
        similar = model.wv.most_similar(phrase, topn=10)
        similar_terms[phrase] = similar
    except KeyError:
        # Term not in vocabulary
        print(f"Phrase '{phrase}' not found in vocabulary.")
        similar_terms[phrase] = []

# Print the results
for phrase, similar in similar_terms.items():
    original_phrase = " ".join(phrase.split("_"))
    print(f"\nPhrase: {original_phrase}")
    for similar_term, similarity in similar:
        similar_term_readable = " ".join(similar_term.split("_"))
        print(f"  {similar_term_readable}: {similarity}")
```
